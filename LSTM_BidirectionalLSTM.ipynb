```python
### meaning of LSTM layer 
import tensorflow as tf
### [32, 10, 8]===> (batch_size, timesteps, number of features)

inputs = tf.random.normal([32, 10, 8])
lstm = tf.keras.layers.LSTM(4)

output = lstm(inputs)

print(output.shape)
```

```python
### return 
lstm = tf.keras.layers.LSTM(4, return_sequences=True, return_state=True)
whole_seq_output, final_memory_state, final_carry_state = lstm(inputs)

whole_seq_output.shape, final_memory_state.shape, final_carry_state.shape
```

```python
inputs = tf.random.normal([32, 10,8])
forward_layer = tf.keras.layers.LSTM(10, return_sequences=True)
backward_layer = tf.keras.layers.LSTM(10, activation='relu', return_sequences=True,go_backwards=True)
bidirectional=tf.keras.layers.Bidirectional(forward_layer, backward_layer=backward_layer)

final_carry_state = bidirectional(inputs)
final_carry_state.shape
```
